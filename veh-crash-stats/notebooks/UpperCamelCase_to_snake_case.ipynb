```python
import re

def convert_upper_camel_to_snake_case(text):
    # Use regular expression to find UpperCamelCase words
    pattern = r'([a-z])([A-Z])'
    snake_case_text = re.sub(pattern, r'\1_\2', text)

    # Convert all characters to lowercase
    snake_case_text = snake_case_text.lower().replace('"','')

    return snake_case_text
```
